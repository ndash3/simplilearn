```python
dataframe_1=pd.DataFrame({'Country_Name':['US','UK','India','Egypt'],'Currency':['Dollar','Pound','Ruppees','Riyal']})
dataframe_1
```

```python
dataframe_2=pd.DataFrame({'Country_Name':['UK','India','Israel'],'IA Names':['MI6','CBI','Mossad']})
dataframe_2
```

```python
dataframe_1.merge(dataframe_2) #Intersection
dataframe_1.merge(dataframe_2,how='left') #Return all rows from left side table will be displayed
dataframe_1.merge(dataframe_2,how='right') #Return all rows from right side table will be displayed
dataframe_1.merge(dataframe_2,how='outer') #Union
```

```python
dataframe_1=pd.DataFrame({'Country_Name':['US','UK','India','Egypt'],'Currency':['Dollar','Pound','Ruppees','Riyal'],'temp':[24,35,22,38]})
dataframe_1
```

```python
dataframe_2=pd.DataFrame({'Country_Name':['UK','India','Israel'],'IA Names':['MI6','CBI','Mossad'],'temp':[22,35,37]})
dataframe_2
```

```python
dataframe_1.merge(dataframe_2,how='left',on='temp')
```

```python
dataframe_1.set_index('Country_Name',inplace=True)
dataframe_2.set_index('Country_Name',inplace=True)
```

```python
dataframe_1.join(dataframe_2,lsuffix='l')
```
